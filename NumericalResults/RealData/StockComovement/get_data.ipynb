```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
# S&P 100
# CommunicationServices = ['CHTR', 'CMCSA', 'DIS', 'GOOG', 'GOOGL', 'META', 'NFLX', 'T', 'TMUS', 'VZ']
CommunicationServices = ['CHTR', 'CMCSA', 'DIS', 'GOOG', 'META', 'NFLX', 'T', 'TMUS', 'VZ']
ConsumerDiscretionary = ['AMZN', 'BKNG', 'F', 'GM', 'HD', 'LOW', 'MCD', 'NKE', 'SBUX', 'TGT', 'TSLA']
ConsumerStaples = ['CL', 'COST', 'KHC', 'KO', 'MDLZ', 'MO', 'PEP', 'PG', 'PM', 'WBA', 'WMT']
Energy = ['COP', 'CVX', 'XOM']
Financials = ['AIG', 'AXP', 'BAC', 'BK', 'BLK', 'BRK-B', 'C', 'COF', 'GS', 'JPM', 'MET', 'MS', 'SCHW', 'USB', 'WFC']
HealthCare =['ABBV', 'ABT', 'AMGN', 'BMY', 'CVS', 'DHR', 'GILD', 'JNJ', 'LLY', 'MDT', 'MRK', 'PFE', 'TMO', 'UNH']
Industrials = ['BA', 'CAT', 'EMR', 'FDX', 'GD', 'GE', 'HON', 'LMT', 'MMM', 'RTX', 'UNP', 'UPS']
InformationTechnology=['AAPL', 'ACN', 'ADBE', 'AVGO', 'CRM', 'CSCO', 'IBM', 'INTC', 'MA', 'MSFT', 'NVDA', 'ORCL', 'PYPL', 'QCOM', 'TXN', 'V']
# Materials = ['DD', 'DOW', 'LIN'] # delete DOW because it starts from 
Materials = ['DD', 'LIN']
RealEstate = ['AMT', 'SPG']
Utilities = ['DUK', 'EXC', 'NEE', 'SO']
print(len(CommunicationServices), len(ConsumerDiscretionary), len(ConsumerStaples), len(Energy),
   len(Financials), len(HealthCare), len(Industrials), len(InformationTechnology),
   len(Materials), len(RealEstate), len(Utilities))

symbol_list = CommunicationServices+ConsumerDiscretionary+ConsumerStaples+Energy+Financials+ \
              HealthCare+Industrials+InformationTechnology+Materials+RealEstate+Utilities

len(symbol_list)
```

```python
def get_close_logret(start, end):
  symbol_data=yf.download(symbol_list, start=start, end=end)
  close=symbol_data['Adj Close']
  close_logret = np.log(close)
  close_logret = close_logret.diff().dropna()
  close_logret = close_logret[symbol_list] # reconstruct in order
  return close_logret
```

```python
data_all = get_close_logret(start="2017-12-30", end="2022-07-01")
```
