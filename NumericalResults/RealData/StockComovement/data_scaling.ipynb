```r
data.c = scale(data.h1.21, center = T, scale = F)
MLE.c = run.mle(data.c)
CLIME.c = run.clime(as.matrix(data.c))
OM.c = OurMethod(as.matrix(data.c))
ANT.c = FastGGM::FastGGM(as.matrix(data.c))
BGGM.c = run.bggm(as.matrix(data.c), prt=TRUE)
GLASSO.c = run.glasso(as.matrix(data.c))
HGL.c = run.hglasso(as.matrix(data.c))
DSGL.c = SILGGM::SILGGM(as.matrix(data.c), method='D-S_GL')
```

```r
data.s = scale(data.h1.21, center = T, scale = T)
MLE.s = run.mle(data.s)
CLIME.s = run.clime(as.matrix(data.s))
OM.s = OurMethod(as.matrix(data.s))
ANT.s = FastGGM::FastGGM(as.matrix(data.s))
BGGM.s = run.bggm(as.matrix(data.s), prt=TRUE)
GLASSO.s = run.glasso(as.matrix(data.s))
HGL.s = run.hglasso(as.matrix(data.s))
DSGL.s = SILGGM::SILGGM(as.matrix(data.s), method='D-S_GL')
```

```r
# Get partial correlation
MLE.c.pc = prec2pc(MLE.c)
CLIME.c.pc = prec2pc(CLIME.c$icov)
OM.c.pc = prec2pc(OM.c$precision.est)
ANT.c.pc = prec2pc(ANT.c$precision)
BGGM.c.pc = prec2pc(BGGM.c$precision)
GLASSO.c.pc = prec2pc(GLASSO.c$wi)
HGL.c.pc = prec2pc(HGL.c$Theta)
DSGL.c.pc = prec2pc(DSGL.c$precision)

MLE.s.pc = prec2pc(MLE.s)
CLIME.s.pc = prec2pc(CLIME.s$icov)
OM.s.pc = prec2pc(OM.s$precision.est)
ANT.s.pc = prec2pc(ANT.s$precision)
BGGM.s.pc = prec2pc(BGGM.s$precision)
GLASSO.s.pc = prec2pc(GLASSO.s$wi)
HGL.s.pc = prec2pc(HGL.s$Theta)
DSGL.s.pc = prec2pc(DSGL.s$precision)
```

```r
plot.scatter<-function(c.pc, s.pc, title,lower=-0.6,upper=0.95){
  df = data.frame(X = c( get.tri(c.pc) ), Y = c( get.tri(s.pc) ))
  ggplot(df, aes(x=X, y=Y)) + geom_point() + geom_abline(intercept = 0, slope = 1, color='red') +
  coord_fixed() + ggtitle(title) +xlim(lower,upper)+ylim(lower,upper)
}
options(repr.plot.width=9, repr.plot.height=5)
# plot.scatter(MLE.c.pc, MLE.s.pc, 'MLE') +
plot.scatter(CLIME.c.pc, CLIME.s.pc, 'CLIME') + xlab(NULL) + ylab('Centered and scaled') +
plot.scatter(ANT.c.pc, ANT.s.pc, 'ANT') + xlab(NULL) + ylab(NULL) +
plot.scatter(BGGM.c.pc, BGGM.s.pc, 'BGGM') + xlab(NULL) + ylab(NULL) +
plot.scatter(GLASSO.c.pc, GLASSO.s.pc, 'GLasso') + xlab(NULL) + ylab(NULL) +
plot.scatter(HGL.c.pc, HGL.s.pc, 'HubGLasso') + xlab('Centered') + ylab('Centered and scaled') +
plot.scatter(DSGL.c.pc, DSGL.s.pc, 'DsGLasso') + xlab('Centered') + ylab(NULL) +
plot.scatter(OM.c.pc, OM.s.pc, 'FLAG') + xlab('Centered') + ylab(NULL) + plot_layout(ncol = 4)
```
